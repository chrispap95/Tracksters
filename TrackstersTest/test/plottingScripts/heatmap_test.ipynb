```python
import numpy as np
```

```python
val_map = {}
```

```python
for node in G5_1.nodes:
    val_map[node] = np.random.rand()
```

```python
values = [val_map.get(node, 0.25) for node in G5_1.nodes()]
```

```python
edges_labels = {}
for edge in G5_1.edges:
    edges_labels[edge] = round(np.random.rand(),1)
```

```python
nx.draw(G5_1, pos=nx.kamada_kawai_layout(G5_1), 
        cmap=plt.get_cmap('inferno'), 
        node_color=values, with_labels=True, 
        font_color='white', font_weight='bold',
        node_size=600)
nx.draw_networkx_edge_labels(
    G5_1, pos=nx.kamada_kawai_layout(G5_1), 
    edge_labels=edges_labels,
    font_color='black'
)
mynodes=nx.draw_networkx_nodes(G5_1, pos=nx.kamada_kawai_layout(G5_1), 
        cmap=plt.get_cmap('inferno'), 
        node_color=values,
        node_size=600)
cbar = plt.colorbar(mynodes,cmap='inferno')
cbar.set_label('Energy [GeV]', rotation=270, fontsize=18, labelpad=40)
```
